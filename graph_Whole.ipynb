```python
#Accuracy difference of all classifiers
import pandas as pd
import matplotlib.pyplot as plt
from decimal import Decimal
import string
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 600

data_file1="D:/PhD/My Papers/Paper2/interdisciplinary science_Revised/graph/UMAPWholeAccuracy.csv"
data1 = pd.read_csv(data_file1, index_col=0)
df1=data1.dropna()
dataF1=df1.iloc[0:2,]
dataF2=df1.iloc[2:4,]
dataF3=df1.iloc[4:6,]
dataF4=df1.iloc[6:8,]
dataF5=df1.iloc[8:10,]
dataF6=df1.iloc[10:12,]
dataF7=df1.iloc[12:14,]
dataFs=[dataF1,dataF2,dataF3,dataF4,dataF5,dataF6, dataF7]
dfsub = pd.DataFrame([])
width = .6
#fig,axs = plt.subplots(6, 1)
Titles=["SVML","NB","kNN","DT","LR","RF","Xgboost"]
for dataF,i,title in zip(dataFs,range(len(dataFs)),Titles):
    diff = -(dataF.diff(axis=0))
    dfsub=dfsub.append(diff.values.tolist())
dfsub=dfsub.dropna(axis=0) 
dfsub.columns =["GSE111629","GSE156994","GSE128235"]
dfsub.index=Titles
plt.figure(figsize=(3, 12))
ax=dfsub.plot.bar(rot=0,align='edge',width=width)
ax.set_ylabel('Acc(Gene)-Acc(Site)')
#ax.set_xlabel('Classifiers')

#ax.legend(ncol = 4)

plt.savefig("D:/PhD/My Papers/Paper2/interdisciplinary science_Revised/graphNew/Fig3.eps", dpi=600)
```
