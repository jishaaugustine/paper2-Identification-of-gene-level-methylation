```python
#Test
from __future__ import division
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import chi2
#from elm import ELMClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,f1_score,matthews_corrcoef,cohen_kappa_score,jaccard_score
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
import lightgbm as lgb
import warnings
from pycm import ConfusionMatrix
warnings.filterwarnings("ignore")
import sys
stdsc = StandardScaler()
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler = MinMaxScaler()
```

```python
import random
random.seed(42)
np.random.seed(42)
def classfiers(X, Y,times,fold):
    #Using six classifiers to do cross validation for X and Y times fold.
    classifier = [SVC(kernel='linear'), GaussianNB(), KNeighborsClassifier(), DecisionTreeClassifier(random_state=10),
                  GradientBoostingClassifier(random_state=10), 
                  LogisticRegression(),
                  RandomForestClassifier(),
                  XGBClassifier(),
                  lgb.LGBMClassifier()
                  ]
    acc_res,f1_res,mcc_res,kappa_res,jaccard_res,csi_res = ([] for i in range(6))
    for clf in classifier:
        #if clf == classifier[5]:
            #X = stdsc.fit_transform(X)
        print(clf)
        each_acc,each_f1,each_mcc,each_kappa,each_jaccard,each_csi = ([] for i in range(6))
        for i in range(times):
            acc_temp,f1_temp,mcc_temp,kappa_temp,jaccard_temp,csi_temp = ([] for i in range(6))
            skf = StratifiedKFold(n_splits=fold, random_state=i, shuffle=True)
            for train_index, test_index in skf.split(X, Y):
                # print('Train: ',train_index,'Test: ',test_index)
                X_train, X_test = X[train_index], X[test_index]
                Y_train, Y_test = Y[train_index], Y[test_index]
                clf.fit(X_train, Y_train)
                #Different metrics
                
                acc = accuracy_score(Y_test, clf.predict(X_test))
                #print("Accuracy..."+str(acc))
                
                f1 = f1_score(Y_test, clf.predict(X_test))
                #print("F1..."+str(f1))
                
                mcc=matthews_corrcoef(Y_test, clf.predict(X_test))
                #print("MCC..."+str(mcc))
                
                
                kappa=cohen_kappa_score(Y_test, clf.predict(X_test))
                #print("kappa..."+str(kappa))
                
                jaccard=jaccard_score(Y_test, clf.predict(X_test))
                #print("jaccard..."+str(jaccard))
                #CSI
               
                cm = ConfusionMatrix(actual_vector=Y_test, predict_vector=clf.predict(X_test))

                #cm.print_matrix()
                csi=cm.CSI
                #print(csi)
                #print(type(csi))
                csiF= 0 if csi is "None" else csi
                #print(csiF)
                ###################
                acc_temp.append(acc)
                f1_temp.append(f1)
                mcc_temp.append(mcc)
                kappa_temp.append(kappa)
                jaccard_temp.append(jaccard)
                csi_temp.append(csiF)
            each_acc.append(np.mean(acc_temp))
            each_f1.append(np.mean(f1_temp))
            each_mcc.append(np.mean(mcc_temp))
            each_kappa.append(np.mean(kappa_temp))
            each_jaccard.append(np.mean(jaccard_temp))
            each_csi.append(np.mean(csi_temp))
            
        acc_res.append(np.mean(each_acc))
        f1_res.append(np.mean(each_f1))
        mcc_res.append(np.mean(each_mcc))
        kappa_res.append(np.mean(each_kappa))
        jaccard_res.append(np.mean(each_jaccard)) 
        csi_res.append(np.mean(each_csi))
    return {"acc":acc_res,"f1":f1_res,"mcc":mcc_res,"kappa":kappa_res,"jaccard":jaccard_res,"csi":csi_res}

def IFS_validation(X, Y, times,fold,tag):
    feature_order = X.columns
    list_gene = []
    result={}
    result_acc,result_f1,result_mcc,result_kappa,result_jaccard,result_csi = ([] for i in range(6))
    print("feature_number:")

    result=classfiers(X.values, Y,times,fold)
    result_acc.append(result["acc"])
    result_f1.append(result["f1"])
    result_mcc.append(result["mcc"])
    result_kappa.append(result["kappa"])
    result_jaccard.append(result["jaccard"])
    result_csi.append(result["csi"])
        
    print("saving acc")    
    score = pd.DataFrame(result_acc)
    col = ['SVM', 'NBayes', 'KNN', 'DTree', 'GBDT','LogisticRegression', 'RF', 'xgboost','lightGBM']
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag + "_acc_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    
    print("saving f1")
    score = pd.DataFrame(result_f1)
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag  + "_f1_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    print("saving mcc")
    score = pd.DataFrame(result_mcc)
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag  + "_mcc_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    print("saving kappa")
    score = pd.DataFrame(result_kappa)
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag  + "_kappa_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    print("saving jaccard")
    score = pd.DataFrame(result_jaccard)
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag + "_jaccard_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    print("saving csi")
    score = pd.DataFrame(result_csi)
    score.columns = col
    #score.index = list(map(lambda x: ','.join(x), list_gene))
    score.to_csv(tag  + "_csi_" + str(times) + "_runs_" + str(fold) + "_fold.csv")
    
    
    
    


```

```python
import random
random.seed(42)
np.random.seed(42)
def select_feature(code,data,label,gene,times=20,fold=5,cnt=100):
    if gene:
        geneOrResidue="gene"
    else:
        geneOrResidue="site"

    tag = code + "_" + geneOrResidue + "_" + model_type # Prefix for result file
    #methods
    IFS_validation(data, label,times,fold,tag)
    
def load_data(filename, labelname):
    data = pd.read_table(filename, index_col=0)
    label = labelname
    return data, label

```
